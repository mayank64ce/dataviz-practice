## Summary

| Concept | What to remember |
|---------|------------------|
| `rcParams` | Global defaults dict — set once, applies everywhere |
| `plt.rcParams.update(dict)` | Apply a whole style dict at once |
| `plt.rc_context(dict)` | Temporary override for a single figure |
| `.mplstyle` file | Ship with your project; load with `plt.style.use(...)` |
| `mpl.rcdefaults()` | Reset everything to matplotlib defaults |
