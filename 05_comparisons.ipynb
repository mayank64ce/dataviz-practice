## Summary

| Technique | Key point |
|-----------|----------|
| Highlight your model | Vivid color for yours, `#BBBBBB` gray for baselines |
| Error bars | `yerr=`, `capsize=4`, `ecolor='#555555'` |
| Grouped bars | Manual `x + offset * bar_width` positioning |
| Horizontal bars | Use `barh()` for 6+ models or long names; sort descending |
| Lollipop | `hlines` + `scatter` — less ink, same information |
| Y-axis start | Don't always start at 0 — zoom in to the relevant range |
